```python
# Regular EDA and plotting libraries
import numpy as np # np is short for numpy
import pandas as pd # pandas is so commonly used, it's shortened to pd 
import matplotlib.pyplot as plt
## Models
from sklearn.ensemble import RandomForestClassifier
```

```python
 cancer = pd.read_csv("/content/drive/MyDrive/DM/can.csv") 
 cancer.head() # (rows, columns)
```

```python
cancer['diagnosis'] = cancer['diagnosis'].astype(int)
```

```python
cancer.head()
```

```python
cancer.isna().sum()
```

```python
# Fill missing values with Scikit Learn
from sklearn.impute import SimpleImputer 
from sklearn.compose import ColumnTransformer
#Fill categorical values with 'missing' & numerical values with mean
num_imputer=SimpleImputer(strategy="mean")
#Define columns
num_features = ["mean_radius","mean_texture","mean_perimeter","mean_area","mean_smoothness","diagnosis"]
imputer=ColumnTransformer([ ("num_imputer",num_imputer,num_features)])
# Transformed the Data
filled_X = imputer.fit_transform(cancer) 
filled_X
```

```python
cancer = pd.DataFrame(filled_X,columns=["mean_radius","mean_texture","mean_perimeter","mean_area","mean_smoothness","diagnosis"])
cancer.head(10)
```

```python
cancer.isna().sum()
```

```python
#Creating X and y
X=cancer.drop("diagnosis",axis=1) 
y=cancer["diagnosis"]
```

```python
np.random.seed(60)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

model = RandomForestClassifier() 
model.fit(X_train,y_train) 
model.score(X_test,y_test)
```

```python
y_preds = model.predict(X_test)
```

```python
output = pd.DataFrame({"mean_texture":X_test.mean_texture,'Cancer?':y_preds}) 
output
```
